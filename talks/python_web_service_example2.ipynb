```python
from owslib.wfs import WebFeatureService 


# connect to geoserver-rc WFS service
serverURL = 'http://geoserver-rc.aodn.org.au/geoserver/ows'
wfs = WebFeatureService(serverURL)

# select a feature type and get its metadata
ftype = 'imos:anmn_nrs_bgc_chemistry_data'
fmeta = wfs.contents[ftype]
```

```python
# Look at some basic metadata 
# (not much there in this case, but there could be)
print fmeta.title
for attrib in ['title', 'abstract', 'keywords', 'boundingBoxWGS84', 'crsOptions']:
    print '  %s: %s' % (attrib, getattr(fmeta, attrib)) 
```

```python
# request data within a bounding box
bbox = (130., -43., 153., -40.)
response = wfs.getfeature(typename=[ftype], bbox=bbox, outputFormat='csv')

# load result into dataframe
import pandas as pd
df = pd.read_csv(response)
df.info()
```
